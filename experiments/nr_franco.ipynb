## NR-FRanCo
In this notebook the experiments for NR-FRanCofor four different PDEs specified in the paper can be performed.<br>
1. $exp(at + x)$ solves $u_t=au=au_x$
2. $cos(at - x)$ solves $u_t=-au_x$
3. $-sin(at - x)$ solves $u_t=-au_x$
4. $(at + x)exp(b t)$ solves $u_t=au_x + bu$
