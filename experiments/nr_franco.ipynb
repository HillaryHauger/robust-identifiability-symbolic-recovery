## NR-FRanCo
In this notebook the experiments for NR-FRanCofor for four different PDEs specified in the paper can be performed.<br>
Additionally, here we consider the uniquness for G up to the third derivative of x. <br>
1. $exp(at + x)$ solves $u_t=au=au_x$
2. $cos(at - x)$ solves $u_t=-au_x$
3. $-sin(at - x)$ solves $u_t=-au_x$
4. $(at + x)exp(b t)$ solves $u_t=au_x + bu$
